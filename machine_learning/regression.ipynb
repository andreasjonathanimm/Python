# Dependencies Import

```python
import sklearn
import pandas as pd
import numpy as np

print("sklearn version: ", sklearn.__version__)
print("pandas version: ", pd.__version__)
print("numpy version: ", np.__version__)
```

# Importing data

```python
data = pd.read_csv('contents/Housing.csv')
data.head()
```

# Splitting Dataset

```python
from sklearn import model_selection

x_train, x_test, y_train, y_test = model_selection.train_test_split(data, data['price'], test_size=0.3, random_state=0)

x_train.head()
```

# Preprocessing Dataset

  Categorical Value

```python
category_columns = ['mainroad', 'guestroom', 'basement', 'hotwaterheating', 'airconditioning', 'prefarea', 'furnishingstatus']
for col in category_columns:
    print(col, " : ", data[col].unique())

from sklearn import preprocessing
category_encoder = preprocessing.OneHotEncoder()
categorical_x_train = x_train[category_columns]
categorical_x_test = x_test[category_columns]
categorical_y_train = y_train
categorical_y_test = y_test

categorical_x_train = category_encoder.fit_transform(categorical_x_train).toarray()
categorical_x_test = category_encoder.transform(categorical_x_test).toarray()
categorical_y_train = categorical_y_train.values.reshape(-1, 1)
categorical_y_test = categorical_y_test.values.reshape(-1, 1)

categorical_x_train.shape
```
